```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
pd.options.display.max_columns = 999
plt.style.use('default')
```

```python
df= pd.read_csv("all.csv", sep=',', skiprows=9, index_col=0)
df=df.iloc[:,6:-5]
df=df.fillna(0)
```

## Cumulative sum

```python
def cumulativesum(dfplot,countries): 
    cumsum=dfplot.cumsum(axis=1)
    cumsum=cumsum.transpose()
    fig,ax = plt.subplots(figsize=(16, 5))
    for i in countries: 
        ax.plot(cumsum.index, cumsum[i],label=i)
    ax.legend(loc='upper center', bbox_to_anchor=(0.2, 0.9),ncol=3, fancybox=True, shadow=True)
    plt.xticks(rotation=90)
    ax.set_title('Arms accumulation', fontsize=10)
    ax.set_xlabel('Year', fontsize=15)
    ax.set_ylabel('Value', fontsize=15)
    plt.show()    
```

```python
cumulativesum(df,['India','Pakistan','Iran','Iraq','Finland'])
```

## Different suppliers

```python
fin= pd.read_csv("fin.csv", sep=',', skiprows=9, index_col=0)
```

```python
fin=fin.iloc[:,7:-5]
fin=fin.transpose()
fin['Soviet Union']=fin['Soviet Union'].fillna(fin['Russia'])
fin=fin.drop('Russia',axis=1)
fin=fin.transpose()
fin=fin.fillna(0)
```

```python
cumulativesum(fin,list(fin.index[:-1]))
```

```python
def percentages(dfplot,countries): 
    cumsum=dfplot.cumsum(axis=1)
    cumsum=cumsum.fillna(0)
    cumsum=cumsum.loc[:,:].div(cumsum.iloc[-1][:])
    cumsum=cumsum.transpose()
    fig,ax = plt.subplots(figsize=(16, 5))
    for i in countries: 
        ax.plot(cumsum.index, cumsum[i],label=i)
    ax.legend(loc='upper center', bbox_to_anchor=(0.9, 0.9),ncol=3, fancybox=True, shadow=True)
    plt.xticks(rotation=90)
    ax.set_title('Arms accumulation', fontsize=10)
    ax.set_xlabel('Year', fontsize=15)
    ax.set_ylabel('Value', fontsize=15)
    plt.show() 
```

```python
percentages(fin,list(fin.index[:-1]))
```
